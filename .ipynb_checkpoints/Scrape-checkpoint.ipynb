```python
#import dependencies 
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
url="http://en.wikipedia.org/wiki/List_of_Nobel_laureates"
page=requests.get(url)
soup=BeautifulSoup(page.content, "html.parser")
soup
```

```python
#Create empty dataframe
columns=[["Year", "Category", "Name", "Nationality", "Age","Known for", "Wikipedia Page"]]
Nobel_winners=pd.DataFrame(columns=columns)
Nobel_winners
```

```python
#find table

nobel_table=soup.find('table',class_='wikitable sortable')

#find every row 
rows=nobel_table.find_all('tr')
```

```python
print(nobel_table.find('tr'))
```

```python
for row in range (1,len(rows)+1):
    values=row.find_all('td')
    for item in range (1,8)
        item.split(";")
```
